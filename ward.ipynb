```python
from sklearn.datasets import make_blobs # 데이터셋 생성을 위한 라이브러리
from scipy.spatial.distance import pdist # 점들 사이의 거리를 계산하기 위한 함수
from scipy.cluster.hierarchy import linkage # 계층적 군집화를 위한 함수
from scipy.cluster.hierarchy import dendrogram # 계층적 군집화 결과를 시각화하기 위한 함수
import numpy as np # 과학 계산을 위한 라이브러리
from matplotlib import pyplot as plt # 데이터 시각화를 위한 라이브러리

# 가상 데이터 생성: 샘플 수 25개, 특성 수 2개, 중심 3개, 클러스터 표준편차 1.5로 설정
X, Y = make_blobs(n_samples=25, n_features=2, centers=3, cluster_std=1.5)
colormap = np.array(['r', 'g', 'b']) # 각 클러스터별 색상 설정
plt.scatter(X[:,0], X[:,1], c=colormap[Y]) # 생성된 데이터 포인트 시각화
plt.title('Generated Data') # 그래프 제목 설정
plt.show() # 그래프 출력

# 유클리드 거리를 사용하여 데이터 포인트들 사이의 거리 행렬 계산
Xdist = pdist(X, metric='euclidean')
# 'ward' 방법을 사용하여 계층적 군집화 수행
Z = linkage(Xdist, method='ward')
# 계층적 군집화 결과를 가지고 덴드로그램 생성 및 시각화
Zd = dendrogram(Z)
plt.title('Dendrogram') # 덴드로그램 제목 설정
plt.show() # 덴드로그램 출력
```
