```python
from sklearn import datasets
import pandas as pd

# Iris 데이터셋 로드
iris = datasets.load_iris()

# 레이블 데이터프레임 생성 및 컬럼 이름 설정
labels = pd.DataFrame(iris.target)
labels.columns = ['labels']

# 특성 데이터프레임 생성 및 컬럼 이름 설정
data = pd.DataFrame(iris.data)
data.columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']

# 레이블을 특성 데이터프레임에 추가
data = pd.concat([data, labels], axis=1)

# 특성 데이터프레임에서 특성만 추출
feature = data[['Sepal length', 'Sepal width', 'Petal length', 'Petal width']]

import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# t-SNE 모델 생성
model = TSNE(learning_rate=100, perplexity=30)

# t-SNE를 사용하여 특성 데이터를 2차원으로 변환
transformed = model.fit_transform(feature)

# 변환된 데이터에서 각각의 차원 값 추출
xs = transformed[:, 0]
ys = transformed[:, 1]

# 산점도 플로팅
plt.scatter(xs, ys, c=labels['labels'])
plt.show()
```
