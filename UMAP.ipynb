```python
import matplotlib.pyplot as plt
import umap
from sklearn.datasets import load_digits

# 손글씨 숫자 데이터셋 로드
digits = load_digits()

# UMAP를 사용하여 차원 축소
embedding = umap.UMAP(n_neighbors=5, min_dist=0.3, metric='correlation').fit_transform(digits.data)

# 축소된 데이터를 플로팅
plt.scatter(embedding[:, 0], embedding[:, 1])  # x축과 y축에 대한 산점도 플로팅
plt.gca().set_aspect('equal', 'datalim')  # 가로축과 세로축의 비율을 같게 설정하여 데이터를 보다 정확하게 표현
plt.title('UMAP로 차원 축소된 데이터셋', fontsize=24)  # 그래프 제목 설정
plt.show()  # 그래프 출력
```
