```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 데이터 생성
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 데이터 시각화
plt.scatter(X[:, 0], X[:, 1])  # X의 첫 번째 열을 x좌표로, 두 번째 열을 y좌표로 사용하여 산점도 플롯
plt.show()

# 각 클러스터 개수에 따른 SSE(Sum of Squared Errors) 계산
SSEs = []
for i in range(1, 11):  # 클러스터 개수를 1부터 10까지 변화시키면서
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)  # KMeans 클러스터링 수행
    kmeans.fit_predict(X)  # 클러스터 할당
    SSEs.append(kmeans.inertia_)  # SSE 값 저장

# 각 클러스터 개수에 따른 SSE 시각화
plt.plot(range(1, 11), SSEs, marker='o')  # 클러스터 개수에 따른 SSE 그래프 플롯
plt.show()
```
