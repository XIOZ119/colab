```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import numpy as np

np.random.seed(1) # 난수 생성 시드 설정
# 데이터 생성
X = np.random.rand(40,1)**2
y = (10-1./(X.ravel()+0.1)) + np.random.randn(40)

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

X_test = np.linspace(-0.1, 1.1, 500).reshape(-1,1) # 테스트용 데이터 생성
fig = plt.figure(figsize=(12,10)) # 그래프 그리기용 subplot 생성

# 다항 회귀 시각화
for i, degree in enumerate([1,3,5,10], start=1):
  ax = fig.add_subplot(2,2,i)
  ax.scatter(X.ravel(), y, s=15) # 실제 데이터 산점도 플롯
  # 다항 회귀 모델 생성하여 학습하고 테스트 데이터에 대한 예측 수행
  y_test = make_pipeline(PolynomialFeatures(degree), LinearRegression()).fit(X,y).predict(X_test)
  # 회귀 결과 그래프 플롯
  ax.plot(X_test.ravel(), y_test, label='degree={0}'.format(degree))
  ax.set_xlim(-0.1, 1.0) #x축 범위 설정
  ax.set_ylim(-2, 12) #y축 범위 설정
  ax.legend(loc='best')
```
