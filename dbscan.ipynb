```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np
from matplotlib import pyplot as plt

# 반달 모양 데이터 생성
X, Y = make_moons(n_samples=1000, noise=0.05)

# 반달 모양 데이터 시각화
plt.title('Half moons')
plt.scatter(X[:,0], X[:,1])
plt.show()

# DBSCAN 알고리즘을 이용한 클러스터링
dbs = DBSCAN(eps=0.1)  # eps 값 설정
Z = dbs.fit_predict(X)  # 클러스터링 수행 및 클러스터 할당

# 클러스터링 결과 시각화
colormap = np.array(['r', 'b'])  # 클러스터링 결과에 따른 색 지정
plt.scatter(X[:,0], X[:,1], c=colormap[Z])  # 클러스터에 따라 다른 색으로 산점도 플롯
plt.title('DBSCAN for half moons')
plt.show()
```
