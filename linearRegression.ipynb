```python
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt

# 데이터 포인트 생성
data = np.array([[30,12], [150, 25], [300, 25], [400, 48], [130, 21], [240, 33], [350, 46], [200, 41], [110, 23], [190, 32], [120, 24], [130, 19], [270, 37], [255, 24]])

# 산점도 그리기
plt.scatter(data[:, 0], data[:, 1])
plt.title("Linear Regression")
plt.xlabel("Delivery Distance")
plt.ylabel("Delievery Time")
plt.axis([0, 480, 0, 50])

x = data[:, 0].reshape(-1, 1) #입력 데이터
y = data[:, 1].reshape(-1, 1) #출력 데이터

# 선형 회귀 모델 생성
model = LinearRegression()
model.fit(x, y) #모델 학습

y_pred = model.predict(x) #예측값 계산
# 회귀선 그리기
plt.plot(x, y_pred, color='r')
plt.show()
```
