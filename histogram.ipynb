```python
import numpy as np
from matplotlib import pyplot as plt

# 정규 분포를 따르는 데이터 생성
X1 = np.random.randn(1000)  # 평균 0, 표준편차 1의 정규 분포를 따르는 데이터 생성
X2 = 10 + np.random.randn(1000)  # 평균 10, 표준편차 1의 정규 분포를 따르는 데이터 생성

# 히스토그램 시각화
plt.figure(figsize=(10, 6))  # 플롯 크기 설정
plt.hist(X1, bins=20, alpha=0.4)  # X1의 히스토그램 플롯
plt.hist(X2, bins=20, alpha=0.4)  # X2의 히스토그램 플롯
plt.show()
```
